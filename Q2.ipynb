```python
import spacy

# Load English NLP model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "Chris met Alex at Apple headquarters in California. He told him about the new iPhone launch."

# Process text
doc = nlp(text)

# ---- Named Entity Recognition ----
print("Named Entities:")
for ent in doc.ents:
    print(f"{ent.text} → {ent.label_}")

# ---- Pronoun Ambiguity Check ----
pronouns = {"he", "him", "she", "her", "they", "them"}
found = [token.text for token in doc if token.text.lower() in pronouns]

if found:
    print("\n⚠️ Warning: Possible pronoun ambiguity detected!")
    print("Pronouns found:", ", ".join(found))
else:
    print("\n✅ No pronoun ambiguity detected.")
```
